```python
from sklearn import datasets #importing all datasets
from sklearn.linear_model import LinearRegression #importing the linear regression model we want a supervised learning
from sklearn.metrics import r2_score, mean_squared_error #importing the metrics we want to use
from sklearn.model_selection import train_test_split #importing the train test split
import matplotlib.pyplot as plt #to draw the graph
```

```python
iris_data = datasets.load_iris() #loading iris dataset
X = iris_data.data[:,2]
y = iris_data.target
y_target = iris_data.target_names
X_reshaped = X.reshape(-1,1) #used to convert 1-D array to 2-D array ideal for machine learning models with just one feature -1 lets numpy determine the feature
X_train, X_test, y_train, y_test = train_test_split(X_reshaped,y,test_size=0.2) #splitting the data into train and test
model_selection = LinearRegression()
model_selection.fit(X_train, y_train) #training the model
y_pred = model_selection.predict(X_test) #predicting the model
plt.scatter(X_test,y_test, color='black',alpha=0.6) #to make a scatter plot for test set
plt.plot(X_test,y_pred, color='blue') #to draw a libe of X_test and y_pred to see how accurate our model is
plt.xlabel("Petal Length")
plt.ylabel("Flower type")
plt.yticks(ticks=[0, 1, 2],labels=y_target)
plt.title("Flower type based on petal length")
#plt.legend("To check if linear regression model can predict the flower species based on sepal length of flower")
plt.show()
#Main metrics for showing performance of linear regression model
print("Mean Squared Error: ", mean_squared_error(y_test,y_pred)) #error should be low
print("R2 score: ", r2_score(y_test,y_pred)) #should be 0.7 to 0.8
```
